```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
duomenys = pd.read_csv('spotify.csv', encoding_errors='ignore')
duomenys
```

Laiko konvertavimas iš milisekundžių į minutes:

```python
duomenys['duration_ms'] = duomenys['duration_ms']/60000
duomenys = duomenys.rename(columns = {'duration_ms':'duration_min'})
```

Po konvertavimo, minučių suapvalinimas.

```python
duomenys['duration_min'] = round(duomenys['duration_min'], 2)
duomenys
```

```python
# @title Pervadinimas
duomenys = duomenys.rename(columns={
    'artist': 'atlikėjas',
    'song': 'daina',
    'duration min' : 'trukmė_min',
    'explicit' : 'N18',
    'year': 'metai',
    'popularity': 'žinonumas',
    'danceability': 'kiek smagu šokti',
    'energy': 'intensyvumas',
    'key': 'tonas',
    'loudness': 'garsumas db',
    'mode': 'tonacija',
    'speechiness': 'žodžių kiekis',
    'acousticness': 'akustiškumas',
    'instrumentalness': 'ar yra vokalas?',
    'liveness': 'ar įrašas live?',
    'valence': 'pozityvumas',
    'tempo': 'tempas',
    'genre': 'žanras'
})


```

Pasak DB autoriaus, duomenys sugeneruoti pačios Spotify programėlės.

**Duration** (trukmė) - kiek laiko trunka daina/ ms?

**N18** - ar dainoje yra įžeidžiantys žodžiai, ar dainos gali klausytis vaikams.

**Metai** - dainos išleidimo metai.

**Žinonumas** - kuo aukštesnė reikšmė, tuo populiaresnė buvo daina. P.S. duomenų autorius nenurodo, kokiu būdu nustatatė dainos populiariumą.

**Kiek smagu šokti**? - kiek daina tinkama šokiams, pvz klubuose. Jei vertė yra 0.0 - nelabai tinka šokiams, o 1.0 - labiausiai šokama.

**Intensyvumas** - nurodomas subjektyvumas kūrinio suvokimas, kuomet kūrinys suvokiamas kaip intensyvus, aktyvus. Matuojama  nuo 0.0. iki 1.0.

**Tonas** - Raktas, kuriame yra takelis. Sveikieji skaičiai susieti su žingsniais, naudojant standartinę Pitch Class žymėjimą. Pvz. 0 = C, 1 = C♯/D♭, 2 = D ir pan. Jei raktas nebuvo aptiktas, reikšmė yra -1.

**Garsumas db** - bendras kūrinio garsumas decibelais. Dažniausiai varijuoja tarp -60 ir 0 db. Garsumas siejamas su stiprumu.

**Tonacija** - nurodomas ar kūrinys yra minoras/ mažoras. Muzikoje minoras ir mažoras yra dvi pagrindinės tonacijų (gamos) rūšys, kurios apibrėžia muzikos skambesį. Jos išsiskiria savo emocine nuotaika ir struktūriniais skirtumais.
Mažoras yra pažymėtas 1, o minoras - 0.

**Žodžių kiekis** - aptinkamas ištartų žodžių kiekis įraše. Kuo daugiau žodžių dainoje, tuo reikšmė artimesnė 1.0. Didesnės nei 0.66 reikšmės nurodo tokius kūrinius, kuriuose vyrauja žodžiai, pavyzdžiui, dainuojamoji poezija. Kai reikšmė tarp 0.33 ir 0.66, tai kūriniai, kuriuose yra tiek muzikos, tiek teksto, įskaitant repą. Mažiau nei 0.33 rodo, kad kūriniai reiškia, kad kūriniai beveik vien instrumentiniai.

**Akustiškumas** - nurodo tokias dainas, kurias atliekant buvo naudojami akustiniai instrumentai t. y. instrumentai, kurie natūraliai generuoja garsą be elektroninių stiprintuvų ar garso efektų. Vertė nuo 0.0 iki 1.0. Jei vertė 1.0, tai rodo, kad labai tikėtina, kad daina akustiška.

**Ar yra vokalas** - rodo, ar kūrinyje yra vokalas. Tokie garsai kaip "ooh" ir "aaa" yra laikomi kaip instrumentai šiame kontekste. Kuo vertė artimesnė 1.0, tuo didesnė tikimybė, kad kūrinyje nėra vokalo. Vertės virš 0.5 reprezentuoja vokalinius kūrinius, tačiau kuo didesnė reikšmė, tuo patikimesnė vertė.

**Ar įrašas live?** - Nurodoma, ar įrašas padarytas gyvai su klausytojais. Kuo aukštesnis gyvumo numeris, tuo didesnė tikimybė, kad įrašas darytas gyvai, pavyzdžiui, koncerto metu. 0.8 vertė indikuoja, kad įrašas labai tikėtina atliktas gyvai.

**Pozityvumas** - Matas nuo 0.0 iki 1.0, apibūdinantis kūrinio perteikiamą muzikinį pozityvumą. Aukšto pozityvumo kūriniai skamba pozityviau (pvz., linksmi, linksmi, euforiški), o mažo valentingumo kūriniai - neigiamiau (pvz., liūdni, prislėgti, pikti).

**Tempas** -  muzikos judėjimo greitis, matuojama taktais per minutę (BPM)

**Žanras** - nurodoma, kokio žanro kūrinys, pavyzdžiui, pop, metal ir pan.

```python
# @title Ar dainų trukmė mažėja?
import matplotlib.pyplot as plt
average_duration_by_year = duomenys.groupby('metai')['duration_min'].mean()

plt.figure(figsize=(12, 6))
plt.figure().set_figwidth(12)
plt.plot(average_duration_by_year.index, average_duration_by_year.values)
plt.ylabel('Minutės')
plt.title('Vidutinė dainos trukmė kiekvienais metais')
plt.grid(True)
plt.show()
```

```python
# @title Žanrai
duomenys['Žanrai'] = duomenys['žanras'].str.split(',').str[0]
genre_counts = duomenys['Žanrai'].value_counts()
sorted_genre_counts = genre_counts.sort_values(ascending=False)
print(sorted_genre_counts)

```

```python
# @title Žanrai, %
# kai žanras priskiriamas pagal pirmą tagą
duomenys['first_genre'] = duomenys['žanras'].str.split(',').str[0]
genre_counts = duomenys['first_genre'].value_counts()
total_songs = genre_counts.sum()
threshold = total_songs * 0.02
filtered_genres = {}
other_count = 0
for genre, count in genre_counts.items():
    if count >= threshold:
        filtered_genres[genre] = count
    else:
        other_count += count
if other_count > 0:
    filtered_genres['Kita'] = other_count
filtered_genre_series = pd.Series(filtered_genres)
colors = plt.colormaps.get_cmap('tab20')(range(len(filtered_genre_series)))
plt.figure(figsize=(5, 5))
plt.pie(filtered_genre_series, labels=filtered_genre_series.index, autopct='%1.1f%%', startangle=50)
plt.title('Dainos pagal žanrus')
plt.show()
```
